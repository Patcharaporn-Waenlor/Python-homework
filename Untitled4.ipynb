```python
from datetime import datetime

class DigitClock:
    def __init__(self):
        pass
    def showTime(self):
        now = datetime.now()
        print("เวลาปัจจุบัน : {}".format(now.strftime("%H:%M:%S")))
```

```python
from datetime import date

class calender:
    def __init__(self):
        pass

    def showDate(self):
        today = date.today()
        print("วันนี้คือวันที่ : {}".format(today))
```

```python
clock = DigitClock()
clock.showTime()
calender = calender()
calender.showDate()
```

```python
class FutureClock(DigitClock, Calender):
    def __init__(self):
        pass

    def show(self):
        print("------------------------")
        self.showDate()
        self.showTime()
        print("------------------------")
```

```python
my_clock = FutureClock()
my_clock.show()
```
