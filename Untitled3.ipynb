```python
class Student:
    def __init__(self, student_ID, card, first_name, last_name, age, gender):
        self.student_ID = student_ID
        self.card = card
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.gender = gender

    def showInfo(self):
        print("""
รหัสนักศึกษา: {}
หมายเลขบัตร 13 หลัก: {}
ชื่อ: {}
นามสกุล: {}
อายุ: {}
เพศ: {}
""".format(self.student_ID, self.card, self.first_name, self.last_name, self.age, self.gender))

    def sit(self):
    
        print("{} is sitting !!".format(self.first_name))

    def sleep(self):
        print("{} is sleeping !!".format(self.first_name))

    def eat(self):
        print("{} is eating !!".format(self.first_name))
```

```python
s1 = Student("6812732134", "1339100047100", "พัชราภรณ์", "แหวนหล่อ", 19, "หญิง")
```

```python
s1.showInfo()
s1.sit()
```
