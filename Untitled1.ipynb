```python
class Dog:
    def __init__(self, name, gender, gene, age, weight, color):
        self.name = name
        self.gender = gender
        self.gene = gene
        self.age = age
        self.weight = weight
        self.color = color

    def showInfo(self):
        print("""
ชื่อ: {}
เพศ: {}
สายพันธุ์: {}
อายุ: {}
น้ำหนัก: {}
สี: {}
""".format(self.name, self.gender, self.gene, self.age, self.weight, self.color))

    def sit(self):
        print("{} is sitting !!".format(self.name))

    def sleep(self):
        print("{} is sleeping !!".format(self.name))

    def eat(self):
        print("{} is eating !!".format(self.name))
```

```python
dog_pokpaek = Dog('ป๊อกแป๊ก', 'male', 'โกลเด้น', 3, 80, 'สีขาว')
dog_jumoh = Dog('จุ๊เม๊าะ', 'female', 'ไซบีเรียน ฮัสกี้', 4, 90, 'สีดำ')
```

```python
dog_pokpaek.showInfo()
dog_pokpaek.sit()
dog_pokpaek.eat()
dog_pokpaek.sleep()
```
