```python
class Dog():
    def _init_(self, name, gender, gene, age, weight, color):
        self.name = name
        self.gender = gender
        self.gene = gene
        self.age = age
        self.weight = weight
        self.color = color

    def showInfo(self):
        print("""
        ชื่อ: {}
        เพศ: {}
        สายพันธ์: {}
        อายุ: {}
        นํ้าหนัก: {}
        สี: {}
        """.format(self.name, self.gender, self.gene, self.age, self.weight, self.color))

    def sit(self):
        print("{} is sitting !!".format(self.name))
        
    def sleep(self):
        print("{} is sleeping !!".format(self.name))
        
    def eat(self):
        print("{} is eating !!".format(self.name))
```

```python
# Instance Variable
dog_popkop = Dog('ป๊อกป๊อก', 'male', 'โกลเด้น', 3, 80, 'สีขาว')
dog_jumoh = Dog('จุ๊เม๊าะ', 'female', 'ไซบีเรียน ฮัสกี้', 4, 90, 'สีดำ')
```
